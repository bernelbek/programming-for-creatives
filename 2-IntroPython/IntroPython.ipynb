Теперь попробуем сделать что-то совсем сложное - решить квадратное уравнение! Вспомним, что это уравнение вида $$ax^2+bx+c=0$$
и у него обычно есть два корня: $x_1$ и $x_2$, которые можно найти по формуле
$$
\begin{array}{ll}
x_1 = & \frac{-b-\sqrt{b^2-4ac}}{2a} \\
x_2 = & \frac{-b+\sqrt{b^2-4ac}}{2a} \\
\end{array}
$$

Нас сейчас не очень интересуют сами уравнения, а скорее то, как нам запрограммировать вычисления по формуле выше. Поскольку решение уравения - отдельная мини-задача, которую может понадобится решать много раз, удобно использовать для этого **функцию**, которую мы опишем с помощью команды `def` (от англ. *define*). А для вычисления квадратного корня используем (импортируем) библиотеку `math`:

```python
import math

def solve(a,b,c):
  d = b*b-4*a*c
  x1 = (-b+math.sqrt(d))/(2*a)
  x2 = (-b-math.sqrt(d))/(2*a)
  return x1,x2

def compute(a,b,c,x):
  return a*x*x+b*x+c

x1,x2 = solve(1,2,-3)
print(x1,x2)

print(compute(1,2,-3,x1))
```

## Работаем с черепашкой

Излагать основы алгоритмов удобно, используя какого-то **исполнителя**, который будет наглядно выполнять наши команды. Мы будем использовать **черепашку** - испольнителя, который умеет двигаться вперёд, оставляя за собой след, и поворачиваться на заданный угол по командам `left` и `right`.

Установим библиотеку для работы с черепашкой:

```python
!pip install jturtle
```

Попробуем что-то нарисовать с помощью черепашки. Для этого нужно дать ей серию команд для рисования, и потом вызвать функцию `done`. Для простоты мы импортируем библиотеку с именем `j`, чтобы короче писать.

```python
import jturtle as t

t.forward(90)
t.right(90)
t.forward(50)
t.right(90)
t.forward(90)
t.right(90+60)
t.forward(100)
t.done()
```

Теперь научимся рисовать квадрат. Для этого также используем функцию:

```python
def square(x):
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)
  t.forward(x)
  t.right(90)

square(100)
t.done()
```

И треугольник:

```python
def triangle(x):
  t.forward(x)
  t.right(120)
  t.forward(x)
  t.right(120)
  t.forward(x)
  t.right(120)

triangle(100)
t.done()
```

Из квадратов и треугольника можно нарисовать домик. Важно правда помнить, в каком положении осталась черепашка после рисования каждой из фигур.

```python
def house(x):
  square(x)
  t.forward(x)
  t.right(30)
  triangle(x)
  t.left(30)
  t.right(180)
  t.penup()
  t.forward(x//3)
  t.left(90)
  t.forward(x//3)
  t.pendown()
  square(x//3)

house(100)
t.done()
```
